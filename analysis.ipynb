# what are in demand job role

```python
#for creating plot
sns.countplot(x = "Job Title", data = dataframe, hue = "Job Title", palette ="flare") 
#for stopping the overlapping od labels
plt.xticks(rotation = 90) 
plt.xlabel("Job Title", fontsize = 10) #to increase fontsize in x-axis
plt.ylabel("Count", fontsize = 10) #to increase fontsize in y-axis
plt.title("Number of Roles by Job Titles", fontsize = 17, color = "black") #to define the title
plt.show()
```

# which company offering most in demand job role

```python
sns.countplot(x = "Company Name", data = dataframe, hue = "Company Name", palette = 'dark:#5A9_r') #for creating plot
plt.xticks(rotation = 90) #for stopping the overlapping od labels
plt.xlabel("company Name", fontsize = 10) #to increase fontsize in x-axis
plt.ylabel("Count", fontsize = 10) #to increase fontsize in y-axis
plt.title("Count of Roles by Company Names", fontsize = 17, color = "green") #to define the title
plt.show() #to display the chart
```

# What is the distribution of salary ranges across different job roles?

```python
plt.figure(figsize = (10,4)) #to change the size of the figure
sns.countplot(x = "Job Title", data = dataframe, hue = "Salary Range", palette = "light:#5A9_r", hue_order = ["3-5LPA", "5-8 LPA", "8-12 LPA", "12-20 LPA", "20+ LPA"])
plt.xticks(rotation = 90)
plt.legend(bbox_to_anchor = (1,1)) #bbox is used to keep the legend outside the graph
plt.show()
```

# What is the average experience required for different job types? 


```python
experience_split = dataframe['Experience Required'].str.extract(r'(\d+)\+?-(\d+)?')
dataframe['Experience Min'] = pd.to_numeric(experience_split[0])
dataframe['Experience Max'] = pd.to_numeric(experience_split[1])
dataframe['Experience Max'].fillna(dataframe['Experience Min'], inplace=True)
dataframe['Average Experience'] = dataframe[['Experience Min', 'Experience Max']].mean(axis=1)

plt.figure(figsize=(8, 5))
sns.barplot(data=dataframe, x='Job Type', y='Average Experience', estimator=np.mean, ci=None)
plt.title('Average Experience by Job Type')
plt.tight_layout()
plt.show()
```

# Which salary range has the highest number of job postings? 

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=dataframe, x='Salary Range', order=dataframe['Salary Range'].value_counts().index)
plt.title('Job Postings by Salary Range')
plt.tight_layout()
plt.show()
```

# Which cities have the highest number of job opportunities

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dataframe, x='Job Location', order=dataframe['Job Location'].value_counts().head(10).index)
plt.title('Top Cities with Most Job Opportunities')
plt.tight_layout()
plt.show()

```

# Most Common Job Types by Location 

```python

top_cities = dataframe['Job Location'].value_counts().head(5).index
city_type = dataframe[dataframe['Job Location'].isin(top_cities)].groupby(['Job Location', 'Job Type']).size().reset_index(name='Counts')
plt.figure(figsize=(10, 6))
sns.barplot(data=city_type, x='Job Location', y='Counts', hue='Job Type')
plt.title('Most Common Job Types in Top Cities')
plt.tight_layout()
plt.show()
```

# which job portals are attracting the highest number of applicants?

```python
plt.figure(figsize=(8, 5))
portal_applicants = dataframe.groupby('Job Portal')['Number of Applicants'].sum().sort_values(ascending=False)
sns.barplot(x=portal_applicants.values, y=portal_applicants.index)
plt.title('Applicants per Job Portal')
plt.tight_layout()
plt.show()

```

# Job Titles by Applications 

```python

top_titles = dataframe.groupby('Job Title')['Number of Applicants'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_titles.values, y=top_titles.index)
plt.title('Top 10 Job Titles by Applications Received')
plt.tight_layout()
plt.show()
```
